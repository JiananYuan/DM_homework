```python
import torch
from torch.nn import Conv1d

data = torch.randn(10, 2, 6)  # 10条数据、数据为2通道、特征数为6
print(data)

conv1d = Conv1d(
    in_channels=2, # 需要和输入数据的通道数对其
    out_channels=3, # 输出数据的通道数
    kernel_size=3, # 卷积核大小
)
output = conv1d(data)
```

```python
print(output)
print(output.shape)
```
